## Step 2: Data Preparation


*   Dropping irrelevant columns and rows
*   Identifying duplicated columns
*   Renaming columns
*   Feature Creation
*   Handling missing values



```python
df.head()
```

```python
# Example of dropping columns
#df = df.drop(columns='deck', axis = 1)
```

```python
df = df[[
    'survived',
    'pclass',
    'sex',
    'age',
    'sibsp',
    'parch',
    'fare',
    #'embarked',
    #'class',
    'who',
    #'adult_male',
    #'deck',
    'embark_town',
    #'alone'
    #'alive'
            ]].copy() #copy command knows its a brand new dataframe and not a copy of the previous dataframe

#survived is similar to alive but here we'll consider the column with numeric values
# embarked is similar to embark_town
# Class is similar to pclass, we'll be picking the column with numeric values
# alone is when parch and sibsp is 0
```

```python
df.loc[df.duplicated()] #locating the duplicate values
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
sns.boxplot(df['age'])
```

```python
df['age'].fillna(df['age'].median(), inplace = True)
```

```python
df.isna().sum()
```

```python
df['embark_town'].fillna(method = 'ffill', inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

## Step 3: Feature Understanding

(Univariate analysis)



*   Plotting feature distributions
*   Histograms
*   KDE
*   Boxplot



```python
ax = df['age'].plot(kind='hist',
               bins=20)
ax.set_xlabel('Age')
```

```python
ax = df['age'].plot(kind='kde')
ax.set_xlabel('Age')
```

```python
fig, axs = plt.subplots(3, 3, figsize=(20, 20))

df['survived'].value_counts()\
    .plot(kind='bar', title = 'Survived', ax = axs[0, 0])
axs[0, 0].bar_label(axs[0, 0].containers[0])


df['pclass'].value_counts()\
    .plot(kind='bar', title = 'P_Class', ax = axs[0, 1])
axs[0, 1].bar_label(axs[0, 1].containers[0])


df['sex'].value_counts()\
    .plot(kind='bar', title = 'Sex', ax = axs[0, 2])
axs[0, 2].bar_label(axs[0, 2].containers[0])

df['age']\
    .plot(kind='kde', title='Age', ax = axs[1, 0])

df['sibsp'].value_counts()\
    .plot(kind='bar', title = 'SibSp', ax = axs[1, 1])
axs[1, 1].bar_label(axs[1, 1].containers[0])

df['parch'].value_counts()\
    .plot(kind='bar', title = 'ParCh', ax = axs[1, 2])
axs[1, 2].bar_label(axs[1, 2].containers[0])

df['who'].value_counts()\
    .plot(kind='bar', title = 'Who', ax = axs[2, 0])
axs[2, 0].bar_label(axs[2, 0].containers[0])

df['fare']\
    .plot(kind='kde', title = 'Fare', ax = axs[2, 1])


df['embark_town'].value_counts()\
    .plot(kind='bar', title = 'Embark Town', ax = axs[2, 2])
axs[2, 2].bar_label(axs[2, 2].containers[0])
```

```python
ax = df['survived'].value_counts()\
                    .plot(kind= 'bar')

ax.bar_label(ax.containers[0])
```

## Step 4: Feature Relationships



*   Scatterplot

```python
ax = sns.scatterplot(
                x = 'age',
                y = 'fare',
                hue = 'pclass',
                palette = 'viridis',
                data = df)
ax.set_title('Age vs Fare')
plt.show()
```

People who travelled in Class 1 seem to have higher fares.

There seem to be a lot of Class 1 passengers between the ages 20 to 60

```python
sns.stripplot(x = df['survived'],
              y = df['age'],
             hue = df['who'])
```
